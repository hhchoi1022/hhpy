# Check overhead time for KCT

```python
from astropy.io import fits
import datetime
from astropy.time import Time
import os, glob
from astropy.table import Table
import matplotlib as plt
```

#### Set up directory & image key

```python
homedir = '/home/hhchoi1022/Desktop/overhead_KCT/'
datelist = os.listdir(homedir)
imkey = homedir+datelist[3]+'/*.fts'
imlist = glob.glob(imkey)
```

```python
targetlist = []
filterlist = []
obstimelist = []
imtypelist = []
altlist = []
for image in imlist:
    hdr = fits.getheader(image)
    obstime = Time(hdr['JD'], format = 'jd').datetime
    target = hdr['OBJECT']
    if 'FILTER' in hdr.keys():
        filter_ = hdr['FILTER']
    else:
        filter_ = ''
    if 'ALTITUDE' in hdr.keys():
        altitude = hdr['ALTITUDE']
    else:
        altitude = ''
    
    imtype = hdr['IMAGETYP']
    targetlist.append(target)
    filterlist.append(filter_)
    obstimelist.append(obstime)
    imtypelist.append(imtype)
    altlist.append(altitude)
    
```

```python
tbl = Table()
tbl['image'] = imlist
tbl['obstime'] = obstimelist
tbl['target'] = targetlist
tbl['filter_'] = filterlist
tbl['imtype'] = imtypelist
tbl['altitude'] = altlist
tbl.sort('obstime')
```

#### Calculate delta(time) bew/ subsequent images

```python
timedeltalist = []
for i, jd in enumerate(tbl['obstime']):
    if i == 0:
        timedelta = 0
    else:
        timedelta = (jd - tbl['obstime'][i-1]).seconds
    timedeltalist.append(round(timedelta,1))
        
tbl['timedelta'] = timedeltalist
```

#### Exclude calibration images

```python
tbl = tbl[tbl['imtype'] == 'Light Frame']
```

# Plotting (tmp)

```python
plt.figure(figsize = (10,8), dpi = 300)
plt.ylabel(r'$\Delta$t (subsequent images)')
plt.plot(tbl['timedelta'], c = 'k')
plt.axhline(120, linestyle = '--', c = 'k', linewidth  =0.3)
plt.ylim(-10, 2000)
plt.xlabel('Image # (in order)')
```
